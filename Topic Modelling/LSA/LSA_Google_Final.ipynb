```python
#lowercasing

pro_list = company_df['pros'].tolist()
import re
pro_lower_list = []
for review in pro_list:
    review = review.lower()
    pro_lower_list.append(review)

```

```python
#Tokenisation
pro_tokenised_list = []
for review in pro_lower_list:
    pro_tokenised = review.split(" ")
    pro_tokenised_list.append(pro_tokenised)
# pro_tokenised_list
```

```python
#alphabetical
pro_words_only_list = []
for review in pro_tokenised_list:
    pro_words_only = [w for w in review if re.search('^[a-z]+$',w)]
    pro_words_only_list.append(pro_words_only)
```

```python
#to split conjoined words
import wordninja

import nltk

nltk.download('words')
english_words = set(nltk.corpus.words.words())



pro_no_combined_list = []

for review in pro_no_stop_words_list :
    new_review = []
    for word in review:
        if word not in english_words:
            word_split_list = wordninja.split(word)
            for w in word_split_list:
                new_review.append(w)
            
        new_review.append(word)
    
    pro_no_combined_list.append(new_review)


```

# LSA model for Pros

```python
#alt LSA code ( FINAL)
import gensim
from gensim import corpora, models
from gensim.corpora import Dictionary
from gensim.models.coherencemodel import CoherenceModel
from gensim.matutils import Sparse2Corpus
from prettytable import PrettyTable



# create a dictionary and bag-of-words representation of the reviews
pro_dictionary = corpora.Dictionary(pro_adj_removed_list)
pro_corpus = [pro_dictionary.doc2bow(review) for review in pro_adj_removed_list]

# build the LSA model
pro_lsa_model = models.LsiModel(pro_corpus, num_topics=3, id2word=pro_dictionary)
pro_coherence_model_lsa = CoherenceModel(model=pro_lsa_model, texts=pro_adj_removed_list, dictionary=pro_dictionary, coherence='c_v')
pro_coherence_lsa = pro_coherence_model_lsa.get_coherence()


#code to display words
# get the top words for each topic
pro_topic_words = []
for i, topic in enumerate(pro_lsa_model.show_topics()):
    pro_topic_words.append([word for word, _ in pro_lsa_model.show_topic(i, topn=20)])

# create a dictionary to store the top words for each topic
pro_word_dict = {}
for i in range(len(pro_topic_words)):
    for j, word in enumerate(pro_topic_words[i]):
        if j in pro_word_dict:
            pro_word_dict[j].append(word)
        else:
            pro_word_dict[j] = [word]

# create a PrettyTable and add the topic words as rows
pro_table = PrettyTable(['Topic ' + str(i+1) for i in range(len(pro_topic_words))])
for j in range(len(pro_topic_words[0])):
    row = []
    for i in range(len(pro_topic_words)):
        row.append(pro_word_dict[j][i])
    pro_table.add_row(row)

print(pro_table)



    
```
